## Questions

View the data below, that shows data for $100$ observations. There are two continuous numerical variables and a dependent variable.

```python
# Seed the pseudo-random number generator
np.random.seed(12)

# Sample size
n=100

# Create data
wcc = np.round(np.random.normal(15, 5, n), 0)
crp = (wcc * 2) + np.round(np.random.normal(0, 10, n), 0)

lipase = wcc + crp + np.round(np.random.normal(2, 10, n), 0)

# Pandas dataframe obejct
df = DataFrame(
    {'WCC':wcc,
     'CRP':crp,
     'Lipase':lipase}
)
```

```python
px.scatter_matrix(
    df,
    title='Scatter plot matrix'
)
```

```python
# Target vector and design matrix
y, X = dmatrices('Lipase ~ WCC + CRP', data = df)
```

### Question 1: Create multiple linear regression model with `WCC` and `CRP` as predictors of `Lipase`.

```python
# 1
model = sm.OLS(y, X).fit()

model.summary2()
```

### Question 2: Comment on the *F* statistic.

The *F* statistic is large at $157.3$, with a small *p* value of $0$. There is enough evidence at the $5\%$ level of significance to show that at least one of the indpendent variables is linearly associated with the dependent variable.

### Question 3: Comment on the individual coefficients and their *p* values.

Only the coefficient for `WCC` has a small *p* value. There is enough evidence at the $5\%$ level of significance to show that `WCC` is linearly associated with `Lipase` after correcting for `CRP`.
