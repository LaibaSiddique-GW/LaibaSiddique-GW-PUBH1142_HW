### Question 1: How do you define a list in Python?

A Python list can be created using square bracket notation. The elements are separated by commas. The following code creates a list of integers and assigns it to variable `my_list`. You can also use the `list` function to create a list. 

```python
my_list = [1, 2, 3, 4, 5]
my_list
```

### Question 2: How do you access the third element of a Python list named `my_list`?

The index value can be used to retrieve an element from a list. Since Python is 0 indexed, you would use index 2 to get the third element. 

```python
my_list[2]
```

### Question 3: How can you add the element "new_item" to the end of an existing list named `my_list`?

The `append` method adds an element to the end of a list. The `extend` method can also be used. 

```python
my_list.append("new_item")
my_list
```

### Question 4: How do you create a Python tuple with the items "apple", "banana", and "cherry"?

A Python tuple can be created using parentheses (although they are not strictly required). The elements are separated by commas. The following code creates a tuple containing `"apple"`, `"banana"`, and `"cherry"` and assigns it to variable `my_tuple`. You can also use the `tuple` function to create a tuple. 

```python
my_tuple = ("apple", "banana", "cherry")
my_tuple
```

### Question 5: How do you access the first element of a Python tuple named my_tuple?

The index value can be used to retrieve an element from a tuple. Since Python is 0 indexed, you would use index 0 to get the first element. 

```python
my_tuple[0]
```

### Question 6: How do you define a dictionary in Python with keys "name", "age" and values "John Doe", 30?

The dict function creates a dictionary. However, curly braces can aso be used, alternatively. The key and the value are separated by a colon, and each key-value pair is separated by a comma. The following code creates a dictionary with keys "name", "age" and values "John Doe", 30. 

```python
my_dict = {"name": "John Doe", "age": 30}
my_dict
```

### Question 7: How can you retrieve the value for the key "name" from the dictionary `my_dict`?

The key can be used as an index to retrieve a value. The `get` method can also be used. Where the index notation returns an error if a key is not found, the get method returns nothing instead of an error message.

```python
my_dict.get("name")
```

### Question 8: How can you change the value associated with the key "age" in the dictionary `my_dict` to 35?

You can use indexing by the name of the key as the index, and assign the value to that key's value that way. The code for this is below. 

```python
my_dict["age"] = 35
my_dict.get("age")  
```

### Question 9: How can you add a new key-value pair "city":"New York" to the dictionary `my_dict`?

You can use the key as a new index for the dictionary and assign the value to that new key's value. The code for this is below. 

```python
my_dict["city"] = "New York"
my_dict
```

### Question 10: How do you create a set in Python with the items "apple", "banana", and "cherry"?

A set is generated using curly brackets or the set function. The latter takes a list as argument. The following code creates a set containing `"apple"`, `"banana"`, and `"cherry"` and assigns it to variable `my_set`. 

```python
my_set = {"apple", "banana", "cherry"}
my_set
```

### Question 11: How do you add the item "orange" to an existing set `my_set`?

The `add` method adds an element to a set. It can be used to add "orange" to my_set

```python
my_set.add("orange")
my_set
```

### Question 12: How do you remove the item "banana" from the set `my_set`?

The `remove` method removes an element from a set by using the name of the element, which in this case is "banana".

```python
my_set.remove("banana")
my_set
```

### Question 13: What is the length of a list `my_list` with elements [1, 2, 3, 4, 5]?

### Quesion 14: What will be the result of `my_list * 2` if `my_list` is [1, 2, 3]?

The result would be `[1, 2, 3, 1, 2, 3]` since all the elements in the list were told to repeat twice by doing `my_list + my_list` (since `* 2` is the same thing as addition of the same thing to itself).
