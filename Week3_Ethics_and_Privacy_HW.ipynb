## Misuse of Health Data

An example of the misuse of health data was the Untreated Syphilis Study at Tuskegee, which was conducted from 1932-1972 by the U.S. Public Health Service (USPHS). The study consisited of 600 African American men recruited from Macon County, Alabama by the promise of free medical care, even though the study's true aim was to study the full progression of the disease and the natural progression of untreated Syphilis. Of the 600 men, 399 were men with latent syphilis and 201 were a control group who were free of the disease. The researchers did not collect informed consent from the participants, especially since they lied to the participants by claiming they were treating the participants "for bad blood, a term commonly used in the area at the time to refer to a variety of ailments" (https://www.history.com/articles/the-infamous-40-year-tuskegee-study). Additionally, the study began when there was no known cure for Syphilis. However, even after penicillin became the recommended treatment for Syphilis in 1947 and easily available, the participants were never offered treatment. The USPHS wanted to continue the study until all paticipants had died, autopsies were performed, and all the project data could be analyzed. However, Peter Buxton, a USPHS venereal disease investigator from San Francisco, leaked the story of this case to a reporter friend, who then passed it Jean Heller, a reporter from the Associated Press. Heller reported the story in July 1972, which lead to public outrage, ultimately forcing the study to finally shut down. At the time the study shut down, 28 participants died from the Syphilis, 100 from related complications, at least 40 spouses had be diagnosed with Syphilis, and it had been passes to 19 children at birth. 

There were multiple ethical issues involved with this experient. The issues included not providing the participants with informed consent when being recruited and before the study began, not offering the partipants treatment when it came out and giving them the choice of whether they wanted to be cured or continue participating in the study, not allowing the partipants to withdraw at any time and insisting to research them even after dealth despite the lack of consent, and decieving and somewhat coercing the participants to stay within the study throught the promise of free medical care, despite not actually treating them. 

This study, and/or the ethical issues invlolved with it could have been prevented by a proper IRB process where the IRB would recognise the study as unethical and prevent it from happening, change its procedures to force the researchers to actually comply with their promise of free medical care for the partipants, and once the treatment became available, to force the researchers to offer the participants treactment. The researchers themselves should also have offerered treatment once it became available without needing to be forces by the IRB. Last but not least, the researchers should have properly informed the recruits about the study, gotten proper informed consent from the participants, and allowed the participants to withdraw from the study at any time.

## Breach of Health Data

The largest healthcare data breach in history was the Change Healthcare data breach in February 2024. The breach occured as a ransomware attack on Change Healthcare, a subsidiary of UnitedHealth Group. The breach affected approximately 192.7 million people and the theft of up to 6TB of data, including personal, medical, and billing records, ultimately causing severe, widespread disruption to U.S. pharmacy and insurance systems. The breach led to an ouatage that caused critical payment and claims processing systems to be halted for around two months, resulting in considerable financial hardship for the healthcare providers that relied on Change Healthcare’s services and delaying essential healthcare services. Responsibility for the attack was claimed by the BlackCat/ALPHV ransomware group. Change Healthcare paid the group $22 million in ransom for the deletion of the stolen data. However, the group did not end up paying its affiliate, which retained a copy of the data, distributed it to another group, and demanded additional ransom, which Change Healthcare did not pay. A lawsuit was filed against Change Healthcare by Nebraska Attorney General Mike Hilger, which "alleged cybersecurity failures, such as continued use of outdated IT systems and poor network segmentation... contributed to the scale of the data breach," as well as "a lack of multifactor authentication, and a failure to isolate backup systems from the primary system, which meant the hacker was able to disable both" and also claimed "the prolonged outage, which was not resolved until two months after the attack, during which time prior authorizations were halted for medical care and prescriptions, [delayed] care and [left] patients without essential medications and prescriptions" (https://www.hipaajournal.com/change-healthcare-responding-to-cyberattack/). The lawsuit also focussed on the company's poor response to the data breach, where people had "to wait for almost five months to be informed that their sensitive data had been stolen in the attack" (https://www.hipaajournal.com/change-healthcare-responding-to-cyberattack/). This healthcare data breach placed people at significant risk of harm by way of identity fraud, identity theft, and other misuses of their personal health information. Change Healthcare tried to dismiss the case. However, the company's claims for dismissal were rejected by Lancaster County District Court Judge Susan Strong "finding that the state [Nebraska] had sufficiently alleged all violations of the state’s consumer protection and data privacy laws" by Change Healthcare (https://www.hipaajournal.com/change-healthcare-responding-to-cyberattack/).


This data breach could have been prevented by Change Healthcare more properly de-anonymizing the data, using up to date IT systems, employing regular software and hardware updates to keep the systems the data was stored on (and hence also the data itself) secure, implementing multi-factor authentication on all systems to prevent unauthorized accesses, having proper backup systems, using proper network segmentation, encrypting data and only decrypting it when it had to be used, and consistently monitoring the data and the systems it was stored on to detect and mitigate any breaches.

## Bias in Health Data

An example of bias in health data was the 2019 Optum healthcare risk algorithm case, which showed significant racial bias in algorithmic health management within the company. The goal of the algorithm was to identify high-risk patients who would benefit from extra care management, such as personalized, in-home assistance and more frequent doctor visits. However, it was found to prioritize white patients over sicker Black patients for specialized care programs, especially since at the same calculated risk score, Black patients were significantly sicker than white patients due to an underestimation of their needs. This was because the algorithm used total health care costs incurred per patient as a proxy for their overall health needs rather than illness and the actual health needs of a patient. Additionally, due to systemic, long-term disparities in access to care, Black patients in the dataset often had lower healthcare expenditures than white patients with the same number of chronic conditions since less money was historically spent on Black patients who had the same level of need. The algorithm ultimately incorrectly interpreted lower spending to indicate that Black patients were healthier than equally sick White patients, leading to lower risk scores, and hence, a failure to identify them for extra care. Ultimately, the algorithm's bias reduced the proportion of Black patients receiving extra help by more than half, exacerbating disparities in the healthcare recieved. 
Despite the algorithm not directly using race as a variable in its calculations, it still produced discriminatory results due to using costs as a proxy for need, which reflected existing societal biases. Researchers determined fixing this bias in the dataset would increase the percentage of Black patients receiving additional help from 17.7% to the estimated true value of Black people needing care of 46.5%. The researchers were ultimately able to reduce the racial bias in the algorithm by 84% by changing the algorithm's target from total cost to a direct measure of health needs. 

This bias could have been avoided by Opum using acutal health needs of a patient to determine their level of sickness instead of total costs from the start, Opum also recognising these financial and healthcare inequalities from the start so they would use actual health needs rather than total costs, and Opum using a more unbiased dataset or a dataset whose biases had been corrected to train the algorithm.

An early example of how data was to be anonymized was described in the Health Insurance Portability and Accountability Act (HIPAA). According to HIPAA, "a data set will be considered anonymized if the directory information about each of the subjects has been removed" (https://harvardonline.harvard.edu/blog/anonymity-de-identification-accuracy-data). Directory information includes, but is not limited to, social security number and name. HIPAA claimed there were about 40 fileds that are considered directory information, and removing those data fields would make the data HIPAA compiant. Since this regulation involeves removing removing data that allows patientd to be directly identified, the technique being used is data redaction. Data redaction is a data anonymization technique that involves removing identifiable information entirely.

### Question 2: What is the Havasupai Tribe case about?

In the early 1990s, researchers from the Arizona State University (ASU) approached the Havasupai tribe about collecting samples for a study surrounding the tribe's high prevalence of Type 2 diabetes. The researchers collected blood samples from tribe members, assuring them that the samples would only be used in the diabetes study. The tribe only gave consent for the diabetesstudy. However, the researchers still had blood samples left over after that study, so they wanted to use them for further research. The IRB allowed the researchers to many other studies with the blood samples without getting the consent from the trube for any of the studies. When the Havasupai Tribe discovered this in 2003, they felt betrayed and violated, as the additional research conducted was not only beyond the scope of the initial consent but also challenged the tribe's cultural and spiritual beliefs. Thus, the Havasupai Tribe sued the Arizona Board of Regents and the ASU, alleging lack of informed consent, misuse of blood samples, and violation of their civil rights. This infringed upon the right of the tribe and its people. Thus, this case become a touchstone in discussions about informed consent, genetic research, the rights of indigenous peoples, and the importance of maintaining the delicate balance between scientific advancement and ethical responsibility.

### Question 5: What is the role of an Institutional Review Board (IRB) in research?

Institutional Review Boards (IRBs), also known as Ethics Committees in some countries, play a crucial role in ensuring that ethical standards in research, especially health related research, are upheld. They serve as the gatekeepers of research ethics, protecting the rights and welfare of research participants. They are independent committees composed of diverse members, including scientists, non-scientists, and community members, who review and monitor research involving human subjects. Their primary responsibility is to ensure that such research is conducted ethically and that the rights and welfare of the participants are adequately protected. They review and approve reseach protocols involving human participants before the study can take place. The IRB assesses the study's design, methodology, and procedures, focusing on the risk-benefit ratio, the informed consent process, and the measures in place to protect participants' privacy and confidentiality. They are also tasked with ongoing monitoring of approved studies, reviewing progress reports, and assessing any proposed changes to the research protocol. If a study poses unexpected risks or if there are instances of non-compliance with ethical standards, the IRB has the authority to require modifications or even suspend or terminate the study.

### Question 7: What is a conflict of interest in public health research?

A conflict of interest in public health research refers to a situation where a person or organization involved in the research has a secondary interest that could unduly influence or bias the conduct, findings, or reporting of the research. This secondary interest could be personal, financial, political, academic, or organizational.

### Question 8: What is informed consent in the context of health data collection?

Consent is a fundamental principle in healthcare and research. It is an expression of respect for the autonomy and dignity of individuals, acknowledging their right to control what happens to their bodies and personal information. In the context of health data collection, consent ensures that individuals are aware of and agree to the collection, storage, and use of their health data.
Informed consent goes a step further, requiring that individuals are provided with sufficient information to make an educated decision about their participation. This includes information about the purpose of the data collection, how the data will be used, who will have access to it, the potential risks and benefits, and the individual's right to withdraw their consent at any time.

### Question 10: What is the principle of k-anonymity in the context of de-identification of health data?

K-anonymity in the context of de-identification of health data involves grouping the data in such a way that each group contains at least  k records, thereby making it difficult to single out individuals. For example, instead of reporting individual patients' incomes, patients might be grouped into income brackets, with each bracket containing at least k patients.

### Question 11: What is the main concern about conflicts of interest in public health research?

Conflicts of interest can threaten the credibility and integrity of public health research, as they can lead to biased results, loss of objectivity, and a breach of trust with research participants and the public.

### Question 12: What is the first step in managing conflicts of interest in public health research?

The first step in managing conflicts of interest is transparency. Researchers should disclose any potential conflicts of interest at the outset of a study, including financial ties, personal relationships, or institutional affiliations that could influence their work. This disclosure should be updated throughout the research process and included in any publications or presentations resulting from the research. Transparency allows others to evaluate the potential influence of the conflict and helps maintain trust in the research process.

### Question 13: What is the role of independent oversight in managing conflicts of interest in public health research?

Independent oversight involves having a separate committee review research proposals for potential conflicts, monitor ongoing research, or review findings before publication. Independent oversight provides an additional layer of scrutiny and helps ensure that conflicts of interest do not compromise the research.

### Question 14: What is one strategy for mitigating conflicts of interest in public health research?

One strategy for mitigating conflicts in public health research is avoidance, where in some cases, the best way to manage a conflict of interest is to avoid it altogether. Avoidance might involve choosing not to participate in certain research or severing financial ties with the industry.

### Question 15: What is the potential impact of not properly managing conflicts of interest in public health research?

Not properly managing conflicts of interest can introduce bias into research, affecting the design, conduct, and reporting of research. This can lead to distorted findings and undermine the validity of the research. It can also erode public trust in research. If the public perceives that researchers are influenced by personal interests, they may be less likely to trust research findings. This can also undermine public confidence in research findings and decrease public engagement with health initiatives or interventions. In an era where misinformation can spread rapidly, maintaining public trust in public health research is paramount. Lastly, not properly managing conflicts of interest can damage the integrity of the research process and the reputation of the research community.
